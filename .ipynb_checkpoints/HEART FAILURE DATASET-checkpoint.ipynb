```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# 1. What is the distribution of age among heart failure patients in the dataset
plt.figure(figsize=(8,6))
sns.distplot(df['age'],kde=True)
plt.title("Age Distribution")
plt.xlabel('age')
plt.show()
```

```python
# 2. How does the death rate vary with age
sns.violinplot(data=df,x='DEATH_EVENT',y='age')
```

```python
# 2. How does the death rate vary with age
sns.histplot(data=df,x='DEATH_EVENT',y='age')
```

```python
# 3. What is the percentage of male and female patients in the dataset
df['sex'].value_counts(normalize=True)
```

```python
# 4. How does the platelet count vary among different age groups
sns.histplot(data=df,x='age',y='platelets')
```

```python
#5 Is there correlation  between  creatinine  and sodium levels  in the blood
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)

# ans: yes the correlation is -0.19
```

```python
sns.countplot(x=df['sex'])
plt.xticks([0,1],['Female','Male'])
plt.show()
```

```python
# 6 how does  the prevalence  of high blood pressure differ betwen  male and femal patients
sns.barplot(x='sex',y='high_blood_pressure',data=df)
plt.xticks([1,0],['Male','Female'])

```

```python
# 6. how does  the prevalence  of high blood pressure differ betwen  male and femal patients
sns.countplot(x='sex',hue="high_blood_pressure",data=df)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

```python
#7 what is relationship between smoking habits and the occurence of heart failure?
sns.histplot(data=df,x='smoking')
```
